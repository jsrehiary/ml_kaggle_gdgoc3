```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
anime_df = pd.read_csv("anime_dataset/anime.csv")

# Tampilkan beberapa baris pertama
anime_df.head()
```

# Visualisasi Data di sini


```python
# Pisahkan genre dan ubah menjadi daftar
genre_list = anime_df['Genres'].dropna().str.split(', ')

# Flatten list dan hitung jumlah kemunculan setiap genre
from collections import Counter
genre_count = Counter([genre for sublist in genre_list for genre in sublist])

# Ubah ke DataFrame untuk visualisasi
genre_df = pd.DataFrame(genre_count.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
genre_df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=genre_df.head(10), x="Count", y="Genre", palette="viridis")

plt.title("10 Genre Anime Terpopuler 2020", fontsize=14)
plt.xlabel("Jumlah Anime")
plt.ylabel("Genre")
plt.show()
```
